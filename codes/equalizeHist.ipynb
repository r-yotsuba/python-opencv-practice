## Sample of equalizeHist()

画像のコントラストを上げるための関数  

### Input
- グレースケール画像への変換が必要
  

[参考(wiki)](https://en.wikipedia.org/wiki/Histogram_equalization)

[opencv doc](https://docs.opencv.org/4.4.0/d6/dc7/group__imgproc__hist.html#ga7e54091f0c937d49bf84152a16f76d6e)

```python
import cv2

import numpy as np
from matplotlib import pyplot as plt

imagePath = "images/lena.png"
# imagePath = "images/sample.png"

img = cv2.imread(imagePath, 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(img, cmap="gray")
```

## ヒストグラム平坦化

```python
equ = cv2.equalizeHist(img)

plt.imshow(equ, cmap="gray")
```
