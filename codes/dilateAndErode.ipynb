## Sample of Dilate() and Erode()

- 必要となる前処理：**2値画像化**(全てのが画素は0 or 225)

### 2値化
- cv2.thresholdを使用
- ret, threshold_img = cv2.threshold(INPUT_IMAGE, threshold_value, 255, cv2.THRESH_BINARY)
- threshold_value：閾値(int)

### Dilate
- 膨張を行う
- 255(白色)領域を増やす -> 近傍の中に白が1つでもあれば、ターゲット画素を黒に置き換える
- [openCV doc](https://docs.opencv.org/master/d4/d86/group__imgproc__filter.html#ga4ff0f3318642c4f469d0e11f242f3b6c)

### Erode
- 収縮を行う
- 近傍に黒が1つでもあれば、ターゲット画素を黒に置き換える
- [openCV doc](https://docs.opencv.org/master/d4/d86/group__imgproc__filter.html#gaeb1e0c1033e3f6b891a25d0511362aeb)

```python
import cv2

import numpy as np
from matplotlib import pyplot as plt

imagePath = "lena.png"
# imagePath = "sample.png"

img = cv2.imread(imagePath, 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(img, cmap="gray")
```

## 画像の2値化

- threshold_valuesを基準に画素値を 0 or 255にする

```python
threshold_value = 127

ret, threshold_img = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)   
```

```python
plt.imshow(threshold_img, cmap="gray")
```

## Dilateの実行

```python
kernel = np.ones((5, 5), np.uint8)
dilate = cv2.dilate(threshold_img, kernel)

plt.imshow(dilate, cmap="gray")
```

## Erodeの実行

```python
kernel = np.ones((5, 5), np.uint8)
erode = cv2.erode(threshold_img, kernel)

plt.imshow(erode, cmap="gray")
```
