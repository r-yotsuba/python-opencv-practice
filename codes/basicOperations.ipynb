```python
import cv2

import numpy as np
from matplotlib import pyplot as plt
```

## 画像の読み込みサンプル

```python
imagePath = "lena.png"

# imreadの第二引数
# 0 -> グレー, 1 -> カラー, -1 -> 無変換
img = cv2.imread(imagePath, 0)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #plt.imshowのカラーフォーマットに合わせる


plt.imshow(img)
```
