## Canny

---

Cannyは、画像のエッジ検出を行うメソッド  

This link is online documentation.  
[OpenCV](https://docs.opencv.org/master/dd/d1a/group__imgproc__feature.html#ga04723e007ed888ddf11d9ba04e2232de)  



```python
import cv2

import numpy as np
from matplotlib import pyplot as plt

imagePath = "lena.png"
# imagePath = "sample.png"

img = cv2.imread(imagePath, 0)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img, cmap="gray")
```

```python
canny = cv2.Canny(img, 100, 200)


plt.imshow(canny, cmap="gray")
```
