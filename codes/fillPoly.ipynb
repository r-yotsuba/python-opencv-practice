## Sample of fillPoly()

### fillPoly()
- 画像中に頂点集合で指定した多角形を描画する関数
- 多角形は指定した色で塗りつぶす
- [openCV doc](https://docs.opencv.org/4.4.0/d6/d6e/group__imgproc__draw.html#ga8c69b68fab5f25e2223b6496aa60dad5)

-----
### parameters
- pts : 頂点集合を記述する
- color : 塗りつぶす色
- [参考にした記事](https://note.nkmk.me/python-opencv-draw-function/)


```python
import cv2

import numpy as np
from matplotlib import pyplot as plt

imagePath = "images/lena.png"
# imagePath = "images/sample.png"

img = cv2.imread(imagePath, 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(img, cmap="gray")
```

```python
pts = np.array(((0,0), (100,0), (200,200)))

# color = (255,255,255)
color = (0, 0, 0)

# inputの ptsを []で囲わないとエラーが出る
fillPoly = cv2.fillPoly(img, [pts], color)

plt.imshow(fillPoly, cmap="gray")
```
